**Tuples**

Tuples may be constructed in a number of ways:
*   Using a pair of parentheses to denote the empty tuple: ()
*   Using a trailing comma for a singleton tuple: a, or (a,)
*   Separating items with commas: a, b, c or (a, b, c)
*   Using the tuple() built-in: tuple() or tuple(iterable)

**Note** that it is actually the comma which makes a tuple, not the parentheses. The parentheses are optional, except in the empty tuple case, or when they are needed to avoid syntactic ambiguity. 

For example, f(a, b, c) is a function call with three arguments, while f((a, b, c)) is a function call with a 3-tuple as the sole argument.

Tuples implement all of the common sequence operations.

```python
# Python Tuples --> very similar to list
tuple = (1000,'Doraremon',123.54,'Shin-chan','Kiteretsu')
tinytuple=('Pokemon',56)

print(tuple) #print the tuple 
print(tuple[0]) #print first element of tuple

print(tuple[1:3]) # from index 1 to (3-1), i.e 2 
print(tuple[2:])  # from index 2 to last index

print(tinytuple * 2)  # Prints list two times

print(tuple + tinytuple) # Prints concatenated tuples
```

**issubset(other)**

set <= other
Test whether every element in the set is in other.

set < other
Test whether the set is a proper subset of other, that is, set <= other and set != other.

**issuperset(other)**

set >= other
Test whether every element in other is in the set.

set > other
Test whether the set is a proper superset of other, that is, set >= other and set != other.

**union(others)**

Return a new set with elements from the set and all others.

**intersection(others)**

Return a new set with elements common to the set and all others.

**difference(others)**

Return a new set with elements in the set that are not in the others.

**symmetric_difference(other)**

Return a new set with elements in either the set or other but not both.

```python
#set operations
s1={10,"Seema",30,"Pooja",50,"Keya"}
s2={30,"Pooja"}

if s2.issubset(s1):
   print("s2 is subset of s1")

s3={20,"Gurmeet", 40,"Zeba", 30, "Pooja",60,"radha",50,"Keya"}

if s1.issuperset(s2):
    print("-------superset----------")
    print(s1.issuperset(s2))
   
if s1.intersection(s3):
  print("---------intersection------------")
  print(s1.intersection(s3))

if s1.union(s2):
   print("---------union1------------")
   print(s1.union(s2))

if s1.union(s3):
  print("---------union2------------")
  print(s1.union(s3))

if s1.difference(s3):               # s1 - s3
    print("-------difference----------")
    print(s1.difference(s3))

if s1.symmetric_difference(s3):  # s1-s3 union s3-s1
    print("-------symmetric difference----------")
    print(s1.symmetric_difference(s3))
```

**Write a program to check entered number is palindrome or not**

```python
def isPalindrome(num):
  temp=num
  rev=0
  while num>0:
    d=num % 10     #extract digit at units place 
    rev=rev*10+d   
    num=num//10
  if(temp==rev):
    return True
  else:
    return False

n=int(input('Enter number : '))


if isPalindrome(n):
    print("The number is a palindrome!")
else:
    print("The number isn't a palindrome!")
```

**Why Use Lambda Functions?**

The power of lambda is better shown when you use them as an anonymous function inside another function.

Say you have a function definition that takes one argument, and that argument will be multiplied with an unknown number:
