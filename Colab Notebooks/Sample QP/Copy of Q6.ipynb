```python
#Store in X , Annual income and spending score
X=mdata.iloc[:,[3,4]].values
X
```

```python
 # Apply K Means algorithm
from sklearn.cluster import KMeans

wcss=[]

for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",random_state=2)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

  print(kmeans.labels_)
  print(kmeans.cluster_centers_)

for i in range(1,len(wcss)):
  print(i+1, ' ', wcss[i])

```

```python
#Use the elbow method to find the optimal number of clusters
plt.plot(range(1,11),wcss)
```

```python
# Fit the data and predict outcome
kmeans=KMeans(n_clusters=3 , init="k-means++" , random_state=2 )
y_kmeans=kmeans.fit_predict(X)
```

```python
# Apply Visualisation to display the clusters

plt.scatter(X[y_kmeans == 0,0] , X[y_kmeans == 0,1] , s=100 ,c='r' , label="C1")
plt.scatter(X[y_kmeans == 1,0] , X[y_kmeans == 1,1] , s=100 ,c='pink' , label="C2")
plt.scatter(X[y_kmeans == 2,0] , X[y_kmeans == 2,1] , s=100 ,c='purple' , label="C3")

plt.scatter(kmeans.cluster_centers_[:,0] , kmeans.cluster_centers_[:,1] , s=100 ,c='yellow' , label="Centroid")

plt.legend()
```
