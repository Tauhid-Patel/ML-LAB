#Perform the following on dataset Social_Network_Ads.csv

```python
#
df1=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/Copy of Social_Network_Ads.csv')
df1
```

```python
# Select input features as age and estimated salary. 

X=df1.iloc[:,[2,3]].values
y=df1.iloc[:,4].values
```

```python
# Apply splitting on the dataset as 80% for training and 20% for testing.
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20 ,random_state=0)
```

```python
#Perform feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# Apply Logistic Regression with default parameters 
from sklearn.linear_model import LogisticRegression

logreg=LogisticRegression(random_state=0)

```

```python
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
```

```python
#Display classification report
from sklearn.metrics import classification_report


print(classification_report(y_test, y_pred))
```
